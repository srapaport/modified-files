```python
import pandas as pd
import duckdb
```

```python
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
result = duckdb.sql("""
    SELECT * 
    FROM read_csv_auto('../results/modified_files_old.csv', strict_mode=false, null_padding=true) 
    WHERE LOWER(path) LIKE '%license%' 
    AND LOWER(branch) LIKE '%main%'
""").fetchdf()
```

```python
result[['origin', 'branch', 'path', 'status']]
```

```python
with open('../results/result.pkl', 'wb') as f:
    pickle.dump(result, f)
```
