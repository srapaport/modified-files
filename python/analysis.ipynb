```python
import pandas as pd
import pickle
```

```python
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.DataFrame()
with open('../results/result.pkl', 'rb') as f:
    df = pickle.load(f)
print(f'size of the dataframe: {len(df)}')
```

```python
df.sample(5)
```

```python
result_grade = pd.DataFrame()
with open('../results/grades.pkl', 'rb') as f:
    result_grade = pickle.load(f)
print(f'size of the dataframe: {len(result_grade):,}')
```

```python
result_grade.sample(5)
```

```python
new_grade = result_grade.loc[result_grade['amount_contrib']>1]
```

```python
# Find all rows in df where the origin exists in new_grade
matching_repos = pd.merge(df, result_grade, on='origin', how='inner')

# Display the result
print(f"Found {len(matching_repos)} repositories that exist in both dataframes")
matching_repos.sample(5)
```

```python
repos_with_rel = matching_repos.loc[matching_repos['amount_rel'] > 0]
repos_with_rel.sample(5)
```

```python
repos_modified = matching_repos.loc[matching_repos['status']=="Modified"]
len(repos_modified)
```

```python
repos_modified.to_csv('../results/repos_modified.csv', index=False)

try:
    repos_modified.to_excel('../results/repos_modified.xlsx', index=False, header=True)
    print(f"Saved {len(repos_modified)} to CSV and Excel files")
except Exception as e:
    print(f"Saved {len(repos_modified)} to CSV file only")
    print(f"Excel save failed: {e}")
```

```python
origins_impacted = pd.read_csv("../results/origin_impacted.csv")
```

```python
origins_impacted.sample(5)
```

```python
rev_counts = origins_impacted['rev'].value_counts().reset_index()
rev_counts.columns = ['rev', 'count']

repos_with_impact_count = repos_with_rel.merge(
    rev_counts, 
    left_on='revision', 
    right_on='rev', 
    how='left'
)
repos_with_impact_count['count'] = repos_with_impact_count['count'].fillna(0).astype(int)
repos_with_impact_count = repos_with_impact_count.drop(columns=['rev'])

print(f"Total repos with releases: {len(repos_with_rel)}")
print(f"Repos with impact count > 1: {len(repos_with_impact_count[repos_with_impact_count['count'] > 1])}")
print(f"Maximum impact count: {repos_with_impact_count['count'].max()}")

repos_with_impact_count.sample(5)
```
